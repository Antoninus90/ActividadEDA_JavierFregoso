```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
np.random.seed(42)

horas_sueno = np.random.normal(6.5, 1.2, 1000)

df1 = pd.DataFrame({
    "ID": range(1, 1001),
    "Horas_sueno": horas_sueno
})

df1.head()
```

```python
grupo1 = np.random.normal(8.5, 0.8, 500)
grupo2 = np.random.normal(7.8, 0.9, 500)

df2 = pd.DataFrame({
    "ID": range(1, 1001),
    "Grupo": ["Ejercicio"]*500 + ["No ejercicio"]*500,
    "Calificacion": np.concatenate([grupo1, grupo2])
})

df2.head()
```

```python
df1.describe()
```

```python
df2.groupby("Grupo").describe()
```

```python
plt.hist(df1["Horas_sueno"], bins=30)
plt.title("Horas de sueño")
plt.xlabel("Horas")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.hist(grupo1, bins=30, alpha=0.5, label="Ejercicio")
plt.hist(grupo2, bins=30, alpha=0.5, label="No ejercicio")
plt.legend()
plt.title("Calificaciones")
plt.show()
```

```python
t_stat, p_value = stats.ttest_1samp(df1["Horas_sueno"], 7)
print("p-valor:", p_value)
```

```python
t_stat, p_value = stats.ttest_ind(grupo1, grupo2)
print("p-valor:", p_value)
```

```python
df1.to_csv("caso1.csv", index=False)
df2.to_csv("caso2.csv", index=False)
```

```python
t_stat, p_value = stats.ttest_1samp(df1["Horas_sueno"], 7)

print("Estadístico t:", t_stat)
print("p-valor:", p_value)
```

```python
t_stat, p_value = stats.ttest_ind(grupo1, grupo2)

print("Estadístico t:", t_stat)
print("p-valor:", p_value)
```

```python
df1.to_csv("caso1.csv", index=False)
df2.to_csv("caso2.csv", index=False)
```
