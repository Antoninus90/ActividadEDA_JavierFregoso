## 1. ¿Qué son los contrastes de hipótesis y para qué se utilizan?

En estadística inferencial, el objetivo fundamental es obtener conclusiones sobre una **población** a partir del análisis de una **muestra** representativa. Dentro de este marco, los **contrastes de hipótesis** son una de las herramientas más potentes, ya que permiten tomar decisiones fundamentadas sobre afirmaciones relativas a parámetros poblacionales.

Un **contraste de hipótesis** es un procedimiento estadístico que evalúa si los datos observados en una muestra son compatibles con una afirmación previa sobre la población (hipótesis nula), o si existe evidencia suficiente para rechazarla en favor de una hipótesis alternativa.

### Elementos clave de un contraste de hipótesis

| Elemento | Descripción |
|---|---|
| **Hipótesis nula (H0)** | Afirmación de partida; generalmente expresa que no existe efecto o diferencia |
| **Hipótesis alternativa (H1)** | Lo contrario a H0; lo que el investigador sospecha o quiere demostrar |
| **Nivel de significancia (α)** | Umbral de error aceptado al rechazar H0 (habitualmente α = 0.05) |
| **Estadístico de contraste** | Valor calculado a partir de los datos muestrales |
| **P-valor** | Probabilidad de obtener un resultado igual o más extremo asumiendo H0 verdadera |

**Regla de decisión:** si el p-valor < α, se rechaza H0. En caso contrario, no se puede rechazar H0.

Los contrastes de hipótesis nos permiten pasar de la mera descripción de datos a la **inferencia**: extrapolar conclusiones al conjunto de la población con un grado controlado de incertidumbre.

## 2. Algoritmo general para realizar un contraste de hipótesis

### 2.1. Contraste para una sola población

1. **Definir el problema** e identificar el parámetro de interés (p. ej., la media μ).
2. **Formular las hipótesis:** H0 (valor de referencia) y H1 (alternativa bilateral o unilateral).
3. **Fijar el nivel de significancia** α (habitualmente 0.05).
4. **Seleccionar el estadístico de contraste** adecuado (para medias con n ≥ 30: t de Student).
5. **Recoger los datos** y obtener una muestra representativa.
6. **Calcular el estadístico** de contraste a partir de los datos muestrales.
7. **Calcular el p-valor** asociado al estadístico.
8. **Tomar la decisión:** si p-valor < α → rechazar H0; si no → no rechazar H0.
9. **Interpretar el resultado** en el contexto del problema original.

### 2.2. Contraste para comparar dos poblaciones

1. **Definir el problema** e identificar las dos poblaciones a comparar.
2. **Formular las hipótesis:** H0: μ1 = μ2 (no hay diferencia); H1: μ1 ≠ μ2 (hay diferencia).
3. **Fijar el nivel de significancia** α.
4. **Seleccionar el estadístico** adecuado (para dos medias independientes: t de Student para muestras independientes).
5. **Recoger datos** de ambas muestras de forma independiente.
6. **Calcular el estadístico** comparando las medias muestrales de ambos grupos.
7. **Calcular el p-valor.**
8. **Tomar la decisión:** si p-valor < α → rechazar H0 y concluir que existe diferencia significativa.
9. **Interpretar el resultado** en el contexto del estudio comparativo.

## 3. Escenario 1: Horas de sueño en adultos (una población)

### 3.1. Planteamiento del estudio

**Problema:** Las recomendaciones sanitarias establecen que los adultos deberían dormir una media de **7 horas diarias**. Sin embargo, el ritmo de vida actual hace sospechar que la población adulta duerme, en promedio, menos de lo recomendado.

**Objetivo:** Determinar si la media de horas de sueño de la población adulta difiere significativamente de las 7 horas recomendadas.

**Tipo de información a recoger:** Se requiere una base de datos con al menos 1.000 individuos adultos, registrando el número promedio de horas de sueño por noche en los últimos 30 días (obtenida mediante encuestas o aplicaciones de monitorización).

En este ejercicio, los datos se han **simulado** mediante una distribución normal con media 6.5 horas y desviación típica 1.2 horas, lo que refleja una situación realista de déficit leve de sueño.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

np.random.seed(42)
```

### 3.2. Generación de la base de datos (Caso 1)

```python
# Simulación de 1000 adultos con horas de sueño distribuidas normalmente
horas_sueno = np.random.normal(6.5, 1.2, 1000)

df1 = pd.DataFrame({
    "ID": range(1, 1001),
    "Horas_sueno": horas_sueno
})

print(f"Base de datos Caso 1: {len(df1)} individuos")
df1.head()
```

### 3.3. Estadística descriptiva

```python
print("Estadística descriptiva - Horas de sueño:")
df1["Horas_sueno"].describe().round(2)
```

```python
# Visualización: histograma de horas de sueño
plt.figure(figsize=(8, 5))
plt.hist(df1["Horas_sueno"], bins=30, color='steelblue', edgecolor='white', alpha=0.8)
plt.axvline(x=7, color='red', linestyle='--', linewidth=2, label='Referencia: 7 horas')
plt.axvline(x=df1["Horas_sueno"].mean(), color='orange', linestyle='-', linewidth=2,
            label=f'Media muestral: {df1["Horas_sueno"].mean():.2f} horas')
plt.title("Distribución de Horas de Sueño en Adultos (n=1000)", fontsize=13)
plt.xlabel("Horas de sueño por noche")
plt.ylabel("Frecuencia")
plt.legend()
plt.tight_layout()
plt.show()
```

### 3.4. Contraste de hipótesis: prueba t para una muestra

**Hipótesis:**
- H0: La media de horas de sueño de la población es igual a 7 (μ = 7)
- H1: La media de horas de sueño de la población es diferente de 7 (μ ≠ 7)
- Nivel de significancia: α = 0.05 (contraste bilateral)

```python
# Contraste t para una muestra
t_stat, p_value = stats.ttest_1samp(df1["Horas_sueno"], 7)

print("=== RESULTADOS DEL CONTRASTE DE HIPÓTESIS (Caso 1) ===")
print(f"Estadístico t: {t_stat:.4f}")
print(f"P-valor:       {p_value:.2e}")
print(f"Nivel de significancia (α): 0.05")
print()
if p_value < 0.05:
    print("→ DECISIÓN: Rechazamos H0")
    print("→ CONCLUSIÓN: Existe evidencia estadísticamente significativa de que")
    print("  la media de horas de sueño es diferente de 7 horas.")
    print(f"  La media muestral ({df1['Horas_sueno'].mean():.2f} h) es inferior al valor de referencia (7 h).")
else:
    print("→ DECISIÓN: No se puede rechazar H0")
```

## 4. Escenario 2: Calificaciones según práctica de ejercicio (dos poblaciones)

### 4.1. Planteamiento del estudio

**Problema:** Se sospecha que la práctica regular de ejercicio físico tiene un efecto positivo sobre el rendimiento académico de los estudiantes.

**Objetivo:** Determinar si existe una diferencia estadísticamente significativa en las calificaciones medias entre estudiantes que practican ejercicio regularmente y los que no lo hacen.

**Tipo de información a recoger:** Base de datos con al menos 1.000 estudiantes divididos en dos grupos:
- **Grupo Ejercicio:** practican ejercicio ≥ 3 veces/semana
- **Grupo No Ejercicio:** no realizan actividad física regular

Para cada estudiante se registra su calificación media en el curso (escala 0-10). Los datos se han **simulado**: Grupo Ejercicio (μ=8.5, σ=0.8) y Grupo No Ejercicio (μ=7.8, σ=0.9).

### 4.2. Generación de la base de datos (Caso 2)

```python
# Simulación de dos grupos de 500 estudiantes cada uno
grupo_ejercicio = np.random.normal(8.5, 0.8, 500)
grupo_no_ejercicio = np.random.normal(7.8, 0.9, 500)

df2 = pd.DataFrame({
    "ID": range(1, 1001),
    "Grupo": ["Ejercicio"] * 500 + ["No ejercicio"] * 500,
    "Calificacion": np.concatenate([grupo_ejercicio, grupo_no_ejercicio])
})

print(f"Base de datos Caso 2: {len(df2)} estudiantes")
df2.head()
```

### 4.3. Estadística descriptiva comparativa

```python
print("Estadística descriptiva por grupo:")
df2.groupby("Grupo")["Calificacion"].describe().round(2)
```

```python
# Visualización 1: Histograma comparativo
plt.figure(figsize=(10, 5))
plt.hist(grupo_ejercicio, bins=30, alpha=0.6, label='Ejercicio', color='steelblue')
plt.hist(grupo_no_ejercicio, bins=30, alpha=0.6, label='No ejercicio', color='salmon')
plt.title('Distribución de Calificaciones por Grupo (n=500 cada grupo)', fontsize=13)
plt.xlabel('Calificación')
plt.ylabel('Frecuencia')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Visualización 2: Boxplot comparativo
plt.figure(figsize=(8, 6))
sns.boxplot(x='Grupo', y='Calificacion', data=df2, palette=['steelblue', 'salmon'])
plt.title('Distribución de Calificaciones por Grupo', fontsize=13)
plt.xlabel('Grupo')
plt.ylabel('Calificación')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### 4.4. Contraste de hipótesis: prueba t para dos muestras independientes

**Hipótesis:**
- H0: No existe diferencia en la media de calificaciones entre ambos grupos (μ1 = μ2)
- H1: Existe diferencia en la media de calificaciones entre ambos grupos (μ1 ≠ μ2)
- Nivel de significancia: α = 0.05 (contraste bilateral)

```python
# Contraste t para dos muestras independientes
t_stat2, p_value2 = stats.ttest_ind(grupo_ejercicio, grupo_no_ejercicio)

print("=== RESULTADOS DEL CONTRASTE DE HIPÓTESIS (Caso 2) ===")
print(f"Estadístico t: {t_stat2:.4f}")
print(f"P-valor:       {p_value2:.2e}")
print(f"Nivel de significancia (α): 0.05")
print()
if p_value2 < 0.05:
    print("→ DECISIÓN: Rechazamos H0")
    print("→ CONCLUSIÓN: Existe evidencia estadísticamente significativa de que")
    print("  las calificaciones medias difieren entre los grupos.")
    media_ej = grupo_ejercicio.mean()
    media_no = grupo_no_ejercicio.mean()
    print(f"  Media Ejercicio: {media_ej:.2f} | Media No Ejercicio: {media_no:.2f}")
    print(f"  El grupo que practica ejercicio obtiene calificaciones superiores en promedio.")
else:
    print("→ DECISIÓN: No se puede rechazar H0")
```
